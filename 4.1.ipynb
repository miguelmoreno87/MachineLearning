```python
# Carga el dataset 'California housing' del tema anterior y realiza un modelo de regresión lineal simple de la mediana
# de los valores de las casas (Y) con cada variable numérica (X) disponible en el dataset. Si únicamente fuera posible escoger un sola
# variable predictora X, ¿qué variable del dataset resulta más adecuada para predecir el valor mediano de las casas?
```

```python
import pandas as pd

# Comenzamos cargando el dataset 'California housing'
housing = pd.read_csv('datasets/housing/housing.csv')

# Aspecto del dataframe
housing[:5]
```

```python
# Para realizar la regresión lineal simple
from sklearn import linear_model # Para el ajuste
import numpy as np
import matplotlib.pyplot as plt # Para representación gráfica
from sklearn.metrics import r2_score, mean_squared_error # Para coeficiente de determinación y varianza

# Se extrae la variable dependiente/respuesta
y = housing['median_house_value']
y_rep = y.values.reshape(-1,1) # Vector columna
```

```python
# Caso 1: Regresión lineal con x = Longitud

# Variable independiente/predictora
x1 = housing['longitude']
x_rep1 = x1.values.reshape(-1,1)

# Objeto para la regresión
lm1 = linear_model.LinearRegression()

# Ajuste
lm1.fit(x_rep1,y_rep)

# Valor teórico
y_pred1 = lm1.predict(x_rep1)

print('La pendiente vale:', lm1.coef_[0][0])
print('La ordenada en el origen vale:', lm1.intercept_[0])
print('El coeficiente de determinación vale:', r2_score(y_rep,y_pred1))
print('La varianza vale:', mean_squared_error(y_rep,y_pred1))

plt.scatter(x1,y)
plt.plot(x1,y_pred1,color='r',linewidth=3)
plt.xlabel('Longitud')
plt.ylabel('Precio medio casa')
plt.show()
```
