```python
# Apartado a)
# La variable que no es de tipo numérico y que es de tipo objeto es 'ocean_proximity'. Se usa el método value_counts() para
# consultar sus valores. Se precisa volver a importar el dataframe del archivo csv (ya descargado) como se hizo en el ejercicio 1.

import pandas as pd

path = 'datasets/housing/housing.csv'
housing = pd.read_csv(path)
print(housing['ocean_proximity'].value_counts())
```

```python
# Apartado b)
# Para obtener información de las variables numéricas se usa el método 'describe'

housing.describe()
```

```python
# Para realizar un histograma de cada columna numérica del dataframe se usa la librería pyplot. Con el parámetro 'bins'
# de la función 'hist' se controla cuántas barras hay en el histograma.

import matplotlib.pyplot as plt

figura = plt.gcf()
figura.set_size_inches(30,30)

plt.subplot(331)
plt.hist(housing['longitude'], bins=50)
plt.xlabel('Longitud')
plt.ylabel('n')

plt.subplot(332)
plt.hist(housing['latitude'], bins=50)
plt.xlabel('Latitud')
plt.ylabel('n')

plt.subplot(333)
plt.hist(housing['housing_median_age'], bins=50)
plt.xlabel('Años')
plt.ylabel('n')

plt.subplot(334)
plt.hist(housing['total_rooms'], bins=50)
plt.xlabel('Puertas')
plt.ylabel('n')

plt.subplot(335)
plt.hist(housing['total_bedrooms'], bins=50)
plt.xlabel('Habitaciones')
plt.ylabel('n')

plt.subplot(336)
plt.hist(housing['population'], bins=50)
plt.xlabel('Habitantes')
plt.ylabel('n')

plt.subplot(337)
plt.hist(housing['households'], bins=50)
plt.xlabel('x')
plt.ylabel('n')

plt.subplot(338)
plt.hist(housing['median_income'], bins=50)
plt.xlabel('Ingreso medio')
plt.ylabel('n')

plt.subplot(339)
plt.hist(housing['median_house_value'], bins=50)
plt.xlabel('Valor medio')
plt.ylabel('n')

plt.show()
```
