```python
# Realiza una figura compuesta de varias subfiguras. En particular, realiza el histograma de las alturas simulado mediante normales en la
# parte superior de la figura y los 3 histogramas de las alturas simuladas mediante uniformes en la parte inferior. Var√≠a los colores de cada
# uno de los histogramas.
```

```python
import matplotlib.pyplot as plt
import numpy as np

alturas = np.random.normal(175, 10, 500)
alturas_1_0 = np.random.uniform(164, 175, 100) # 100 valores reales entre 164 y 175 seg√∫n una distribuci√≥n uniforme
alturas_1 = np.append(alturas_1_0, 180) # Se a√±ade 180, un outlier, como valor adicional a alturas_1_0 y se guarda en un nuevo array
alturas_2 = np.random.uniform(175, 185, 100)
alturas_3 = np.random.uniform(170, 185, 100)

plt.figure(figsize=(12,13))

# Primera figura (Ocupa la primera fila y las tres columnas. Esta figura cuenta como tres y por eso el siguiente subplot empieza con id=4)
plt.subplot(2,1,1)
plt.hist(alturas, bins=10, edgecolor='black', color='blue')
plt.title('Distribuci√≥n de 500 alturas simuladas de una normal')
plt.xlabel('Altura (cm)')
plt.ylabel('n')

# 3 figuras siguientes, ocupando la misma fila (la segunda)
plt.subplot(2, 3, 4)
plt.hist(alturas_1, bins=10, edgecolor='black', color='green')
plt.title('Distribuci√≥n alturas 1')
plt.xlabel('Altura (cm)')
plt.ylabel('n')

plt.subplot(2, 3, 5)
plt.hist(alturas_2, bins=10, edgecolor='black', color='purple')
plt.title('Distribuci√≥n alturas 1')
plt.xlabel('Altura (cm)')
plt.ylabel('n')

plt.subplot(2, 3, 6)
plt.hist(alturas_3, bins=10, edgecolor='black', color='red')
plt.title('Distribuci√≥n alturas 1')
plt.xlabel('Altura (cm)')
plt.ylabel('n')

plt.show()
```
