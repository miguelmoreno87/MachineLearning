```python
# Partiendo del diagrama de cajas y bigotes realizado en el tema como ejemplo, aplica los siguientes cambios sobre su formato:
# a) Que los cuerpos del histograma estén mellados (notched)
# b) Que la altura 1 sea de color azul, la altura 2 de color verde y la altura 3 de color naranja.
# c) Que el color de la mediana, la línea vertical y de los bigotes sea morada.
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
alturas_1_0 = np.random.uniform(164, 175, 100) # 100 valores reales entre 164 y 175 según una distribución uniforme
alturas_1 = np.append(alturas_1_0, 180) # Se añade 180, un outlier, como valor adicional a alturas_1_0 y se guarda en un nuevo array
alturas_2 = np.random.uniform(175, 185, 100)
alturas_3 = np.random.uniform(170, 185, 100)

# Se pasa notch=True para cambiar la forma de las cajas y patch_artist=True para dar color a las cajas. Para controlar los colores de la
# mediana, la línea vertical y de los bigotes se usan los parámetros capprops, medianprops y whiskerprops, respectivamente.
box = plt.boxplot([alturas_1, alturas_2, alturas_3], notch=True, patch_artist=True, capprops=dict(color='purple'),
                 medianprops=dict(color='purple'), whiskerprops=dict(color='purple'))
plt.xticks([1,2,3], ['1','2','3'])
plt.ylabel('Alturas (cm)')
plt.xlabel('Grupo')
colors = ['blue', 'green', 'orange']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
plt.show()
```
