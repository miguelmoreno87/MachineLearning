```python
# a) Comprueba la normalidad de una muestra obtenida mediante muestreo aleatorio de una distribución normal con media
# 0.05, desviación típica 0.9 y tamaño 1000
# b) Comprueba la normalidad de una muestra obtenida mediante el muestreo aleatorio de una distribución normal con media 0.05, desviación
# típica 0.9 y tamaño 10000
```

```python
# Apartado a)

import numpy as np
from scipy.stats import shapiro

# Array con 1000 valores aleatorios de una distribución normal.
np.random.seed(seed=11)
muestra = np.random.normal(0.05,0.9,1000)

# Para comprobar la normalidad, como n=1000<5000, se usa la prueba de Shapiro-Wilk. Se usa la función shapiro del módulo scipy.stats.
sh_resultado = shapiro(muestra)
print('Estadístico:', round(sh_resultado.statistic,3), 'p-valor:', round(sh_resultado.pvalue,3))
print('Si p-valor > 0.05 se acepta la hipótesis nula de normalidad de la muestra')
```

```python
# Apartado b)

from scipy.stats import kstest

np.random.seed(seed=11)
muestra2 = np.random.normal(0.05,0.9,10000)

# Como la muestra es de tamaño n=10000>5000 se usa la prueba de Kolmogorov-Smirnov. Se usa la función kstest del módulo scipy.stats.
ks_resultado = kstest(muestra2, 'norm', args=(0.05,0.9))
print('Estadístico:', round(ks_resultado.statistic,3), 'p-valor:', round(ks_resultado.pvalue,3))
print('Si p-valor > 0.05 se acepta la hipótesis nula de normalidad de la muestra')
```

```python
# También se puede comprobar la normalidad mediante una prueba gráfica. Se representan los cuantiles teóricos frente a 
# los cuantiles muestrales y se comprueba si se obtiene o no una recta. Para la representación se usa la librería matplotlib.pyplot.
# Para el cálculo de los cuantiles se usa la librería scipy.stats.

import matplotlib.pyplot as plt
import scipy.stats as stats

# Dimensiones de la figura
plt.figure(figsize=(5,5))

stats.probplot(muestra, dist='norm', plot=plt)
plt.xlabel('Cuantiles teóricos')
plt.ylabel('Cuantiles muestrales')
plt.title('Test visual')
plt.show()
```
